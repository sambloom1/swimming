```python
import pandas as pd

# Specify the path to the Excel file
excel_path = "C:/Users/Sam/OneDrive/Attachments/Samples Tracker Meda.xlsx"

# Read the Excel file into a pandas DataFrame
df = pd.read_excel(excel_path)

# Display the DataFrame
print(df)
```

```python
df = pd.read_excel(excel_path, header=2)
print(df)
```

```python
df = df['Requested by', 'Reason']
```

```python
df = df[['Requested by', 'Reason']]
```

```python
df = df[df['Reason'].notnull()]
```

```python
df.head()
```

```python
df_pivot = df.pivot_table(index='Requested by', columns='Reason', aggfunc='size', fill_value=0)
```

```python
df_pivot.plot(kind='bar', stacked=True)
```

```python
names = ['Erin' , 'Harriet', 'Jess']
df_filter_name = df[df['Requested by'].isin(names)]
```

```python
df_pivot_name = df_filter_name.pivot_table(index='Reason', columns='Requested by', aggfunc='size', fill_value=0)
```

```python
df_pivot_name.plot(kind='bar', stacked=True)
```

```python
import matplotlib.pyplot as plt

# Create the stacked bar chart
ax = df_pivot_name.plot(kind='bar', stacked=True, figsize=(12, 7), colormap='viridis')

# Add a title
ax.set_title('Stacked Bar Chart of Reason vs Requested by', fontsize=16)

# Add x and y labels
ax.set_xlabel('Reason', fontsize=14)
ax.set_ylabel('Count', fontsize=14)

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Move the legend to an unobtrusive location
ax.legend(title='Requested by', bbox_to_anchor=(1.05, 1), loc='upper left')

# Remove the frame around the legend box
ax.legend(frameon=False)

# Remove the top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Show grid lines for easier comparison of bar heights
ax.grid(True)
ax.set_facecolor('orange')
# Display the plot
plt.tight_layout()
plt.show()
```
