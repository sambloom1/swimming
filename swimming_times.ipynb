```python
import pandas as pd

df = pd.read_csv('C:/Users/Sam/Documents/GitHub/swimming/20170306-swimming-times3.csv', encoding = 'ISO-8859-1')

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

filtered_df = df[(df['Event'] == '200 m freestyle')]
filtered_df = filtered_df[(filtered_df['Ranking'] == '1')]

# Print out the DataFrame to check it
print(filtered_df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

filtered_df = df[df['Ranking'] == '1']

# Create a barplot
plt.figure(figsize=(10, 6))
sns.countplot(x='Nationality', data=filtered_df)
plt.title('Number of Wins per Nationality for all events')
plt.xlabel('Nationality')
plt.ylabel('Number of Wins')
plt.xticks(rotation=90)
plt.show()
```
